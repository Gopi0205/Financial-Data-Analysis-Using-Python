```python
import datetime
from pandas_datareader import data
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
# Ensure yfinance overrides pandas_datareader
yf.pdr_override()
```

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
# Fetch the data for Bank of America
BAC = data.get_data_yahoo('BAC', start=start, end=end)
#Citi Group
C = data.get_data_yahoo('C', start=start, end=end)
#Goldman sachs
GS = data.get_data_yahoo('GS', start=start, end=end)
#JPmorgan Chase
JPM = data.get_data_yahoo('JPM', start=start, end=end)
#Morgan Stanley
MS = data.get_data_yahoo('MS', start=start, end=end)
#Wells Fargo
WFC = data.get_data_yahoo('WFC', start=start, end=end)
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

```python
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return']=bank_stocks[tick]['Close'].pct_change()
```
